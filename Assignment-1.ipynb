Write programs to find the root of f(x) = e^−x − x using bisection, false-position method and
Newton-Raphson methods(True root = 0.56714329). Plot the ”true error” in all the cases as a function of
iteration to show which of them converges faster.

```python
def f1(x):
    return np.exp(-x)-x
def df1(x):
    return -1-np.exp(-x)
```

```python
def bisection1(f, up, low, error, itr):
    if f(up)*f(low) < 0:
        e = []
        true_value = 0.56714329
        for i in range(itr):
            if i == itr-1:
                print("MAX iteration have been reached(bisection)")
                return new, e
            new = (up + low)/2
            e.append(np.abs(new - true_value))
            if np.abs(f(new))<error:
                return new, e
            elif f(new) * f(up) < 0:
                low = new
            else:
                up = new
    else:
        print("Wrong Limit Taken")
        return
```

```python
def newton_raph1(f, df, initial, error, itr):
    true_value = 0.56714329
    e = []
    a = initial
    for i in range(itr):
        if i == itr-1:
                print("MAX iteration have been reached(newton)")
                return a, e
        e.append(np.abs(a-true_value))
        a = a - (f(a)/df(a))
        if np.abs(f(a)) < error:
            return a, e
```

```python
def false_pos1(f, up, low, error, itr):
    true_value = 0.56714329
    e = []
    if f(up) * f(low) < 0:
        for i in range(itr):
            if i == itr-1:
                print("MAX iteration have been reached(false)")
                return new, e
            new = (((up*f(low))- (low * f(up)))/ (f(low)-f(up)))
            e.append(np.abs(new - true_value))
            if np.abs(f(new))<error:
                return new, e
            elif f(new) * f(up) < 0:
                low = new
            else:
                up = new
    else:
        print("Wrong limit")
        return
```

```python
x = np.linspace(-2, 2, 100)
y = f1(x)
plt.plot(x, y)
plt.grid(color = 'black')
plt.show()
```

```python
root_bis , e1 = bisection1(f1, 0.5, 1, 10**(-6), 20)
root_false , e2 = false_pos1(f1, 0.5, 1, 10**(-6), 20)
root_newton , e3 = newton_raph1(f1, df1,1, 10**(-6), 20)
```

```python
plt.plot(range(1, len(e1) + 1), e1, label='Bisection Method')
plt.plot(range(1, len(e2) + 1), e2, label='False-position Method')
plt.plot(range(1, len(e3) + 1), e3, label='Newton-Raphson Method')
plt.yscale('log') 
plt.xlabel('Iteration')
plt.ylabel('True Error')
plt.legend()
plt.show()
```

Determine the positive real root of x^20 − 1 using bisection, false-position method and Newton-Raphson methods. Plot the ”approximate error” in all the cases as a function of iteration to show which of them converges faster. Use different initial conditions and show this.

```python
def f2(x):
    return (x**20) - 1
def df2(x):
    return 20* (x**19)
```

In a diatomic molecule NaCl, the interaction potential between the N a+ and Cl− ions can be
modelled as: 
V (r) = −(e^2)/4π0r + V0 exp(−r/r0).

At equilibrium the force between the ions is: f(r) = −dV(r)/dr = −(e^2)/4π0(r^2) +(V0/r0)exp(−r/r0). Find the root of
f(r) = dg(r)/dr = 0 with g(r) = −V (r).

This will give us the equilibrium distance between the ions. Use
V0 = 1.09 × 103 eV , 
r0 = 0.330A◦ and, 
e^2/4π0 = 14.4A◦eV .

```python
def v(r):
    return (14.4/r) - (1.09 *(10**3)) * np.exp(-r/0.330)
def dv(r):
    return -(14.4/ (r**2)) - (3.303 * (10**3)) * np.exp(-r/0.330)
```

```python
x = np.linspace(-10,10,5000)
y = v(x)
plt.plot(x, y)
plt.ylim(-1000, 1000)
plt.xlim(-10, 10)
plt.grid()
plt.show()
```
